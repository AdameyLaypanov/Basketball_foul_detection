```python
import cv2
import os

video_path = '/Users/admlanno/Desktop/Курсовая/processed_vidoes/attacker_foul1.mp4'
output_folder = '/Users/admlanno/Desktop/Курсовая/Frames/attacker_foul1'

# Создайте папку, если она не существует
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)
frame_rate = int(cap.get(cv2.CAP_PROP_FPS))  # Количество кадров в секунду
count = 0
frame_skip = 5  # Сохранять каждый 5-й кадр

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Сохранять только каждый frame_skip кадр
    if count % frame_skip == 0:
        frame_filename = f"{output_folder}/frame_{count:04d}.jpg"
        cv2.imwrite(frame_filename, frame)
    count += 1

cap.release()
print(f"Extracted {count // frame_skip} frames out of {count} total frames.")
```
